```python
import numpy as np
loan=open('loans_500.txt', mode='r')
data=loan.readlines()
def process_data(data):
    for i in range(1, len(data)):
        data[i]=[float(x) for x in data[i].strip().split(',')]
    key=data[0].strip().split(',')
    return key, data[1:]
```

```python

def make_dictionary(key, values):
    data_list=[]
    for value in values:
        records = dict(zip(key,value))
        data_list.append(records)
    return data_list
```

```python
def data_making(data):
    key, values=process_data(data)
    records=make_dictionary(key, values)
    return records
 
```

```python
def calculate_installments(loan_amount, down_payment, interest_rate, time, records):
    amount=loan_amount-down_payment
    rate=interest_rate/12
    installment=(amount*rate*(1+rate)**time)/((1+rate)**time-1)
    return installment
```

```python
def compute_data():
    records=data_making(data)
    for record in records:
        record['emi']=calculate_installments(record['loan_amount'], record['down_payment'], record['interest_rate'], record['duration_months'], record)
    print(records)
loan.close()
```
