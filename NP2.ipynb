```python
import numpy as np

# Set a seed so you get the same 'random' numbers every time you test
np.random.seed(42)

# Generate 500 records
num_records = 500

# Columns: Amount (10k-50k), Rate (3%-15%), Duration (12-72 months), Down Payment (1k-10k)
loan_amounts = np.random.uniform(10000, 5000000, (num_records, 1))
rates = np.random.uniform(3, 15, (num_records, 1))
durations = np.random.choice([12, 24, 36, 48, 60, 72], (num_records, 1)).astype(float)
down_payments = np.random.uniform(1000, 100000, (num_records, 1))

# Combine into one matrix
data = np.hstack([loan_amounts, rates, durations, down_payments])

# Save to file
header = "loan_amount,interest_rate,duration_months,down_payment"
np.savetxt('loans_500.txt', data, delimiter=',', header=header, comments='', fmt='%.2f')

print("Created loans_500.txt with 500 records!")
```

```python
with open('loans_500.txt', 'r') as file:
    sum=0
    lines = file.readlines()
    for line in lines[1: ]:  
        sum=sum+1
print("Total number of records :",sum)        
```
